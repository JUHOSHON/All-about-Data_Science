# DataFrame 값 추가 삭제

```python
iphone_df.loc['iPhone XR'] = ['2018-10-26', 6.1, '3GB', 'iOS 12.0.1', 'Yes']
# 없는 row이름으로 값을 넣으면 새로운 row가 생성됨
```

```python
iphone_df
```

```python
iphone_df['제조사'] = "Apple"
iphone_df
```

```python
iphone_df.drop('iPhone XR', axis = 'index', inplace = True) 
# 판다스에서는 row를 index라고도 부름 : row를 삭제하고 싶다는 말
# inplace = False 는 기존의 dataframe은 건드리지 않겠다는 말
```

```python
iphone_df
```

```python
iphone_df.drop('제조사', axis = 'columns', inplace = True)
iphone_df
```

```python
iphone_df.drop(['iPhone 7', 'iPhone 8', 'iPhone X'], axis = 'index', inplace = True)
iphone_df
```
