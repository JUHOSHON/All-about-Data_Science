좋은 데이터의 기준
1. 완결성 : 필수적인 데이터는 모두 기록되어 있어야함
   완결성은 어떻게 알 수 있을까 : 결측값(NaN)의 유무를 확인
2. 유일성 : 동일한 데이터가 불필요하게 중복되어 있으면 안됨
3. 통일성 : 데이터가 동일한 형식으로 저장되어 있어야 함 데이터 타입, 단위, 포맷
4. 정확성 : 데이터가 정확해야 함(이상점이 없어야 함)

# 데이터 클리닝 : 완결성

### 결측값이 발생한다면?

```python
import pandas as pd
```

```python
df = pd.read_csv('data/attendance.csv', index_col = 0)
df
```

```python
df.isnull() # 결측값이 있는 곳에만 true 표시
```

```python
df.isnull().sum()
```

```python
# 1. 결측값을 그냥 그대로 둔다( 그 자체로도 의미가 있는 경우)
```

```python
# 2. 결측값을 지워버리는 방법
df.dropna()
```

```python
df # 그냥 dropna를 하면 기존 데이터프레임을 건드리지 않음
# inplace = true 필요
```

```python
df.dropna(axis = 'columns') # 결측값 컬럼 삭제
```

```python
# 있는 데이터를 모두 보존하고 싶다면 - 결측값을 대체
df.fillna(0) # 0으로 채워넣음
```

```python
df.mean()
```

```python
df.fillna(df.mean())
```

```python
df.median()
```

```python
df.fillna(df.median())
```
